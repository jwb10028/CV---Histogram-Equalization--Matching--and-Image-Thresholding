```python
import matplotlib.pyplot as plt 
import matplotlib.image as img
import numpy as np
import cv2


p_img = cv2.imread('./people.png', cv2.IMREAD_GRAYSCALE)
a_img = np.array(p_img)
n_img = ((a_img - np.min(a_img)) * (1/(np.max(a_img) - np.min(a_img)) * 255)).astype('uint8')

# df function
def create_pdf(img_in):
    # Initialize a list with 256 zeros to represent the pdf bins
    pdf = [0] * 256
    total_pixels = 0

    # Iterate through each pixel in the image
    for row in img_in:
        for pixel in row:
            # Update the corresponding bin in the pdf
            pdf[pixel] += 1
            total_pixels += 1

    pdf = [count / total_pixels for count in pdf]

    return pdf


# Calculated pdf
pdf = create_pdf(n_img)
print("Pre-Equalization pdf")
plt.plot(pdf, color="red")
plt.show()


# cdf function
def create_cdf(pdf):
    
    # Initialize a list to store the cumulative distribution function
    cdf = [0] * len(pdf)
    cdf[0] = pdf[0]

    # Calculate the cumulative distribution function
    for i in range(1, len(pdf)):
        cdf[i] = cdf[i - 1] + pdf[i]

    # Normalizing
    cdf = [j * 255 / cdf[-1] for j in cdf]

    return cdf


# Calculated cdf
cdf = create_cdf(pdf)
print("Pre-Equalization cdf")
plt.plot(cdf, color="red")
plt.show()

# histogram_equalization function
def histogram_equalization(img_in):
    # pdf & cdf of input img
    pdf = create_pdf(img_in)
    cdf = create_cdf(pdf)
  
    equalized_img = []
    for row in img_in:
        equalized_row = []
        for pixel_value in row:
            equalized_row.append(cdf[pixel_value])
        equalized_img.append(equalized_row)
 
    # update to return equalized image
    return equalized_img

ep_img = histogram_equalization(n_img)
arr = np.array(ep_img)
cv2.imwrite('people_processed.png', arr)

p2_img = cv2.imread('./people_processed.png', cv2.IMREAD_GRAYSCALE)
n2_img = np.array(p2_img)
pdf2 = create_pdf(n2_img)
cdf2 = create_cdf(pdf2)

print("Post-Equalization pdf")
plt.plot(pdf2, color="blue")
plt.show()

print("Post-Equalization cdf")
plt.plot(cdf2, color="blue")
plt.show()

plt.imshow(p2_img, cmap = 'gray', vmin = 0, vmax = 255 )

```
